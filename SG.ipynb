### SG 2023 data processing 

```python
import pandas as pd
```

```python
# Read the Excel file
file_path1 = r"C:\Users\Kitty Wong\05 KW\Data\SG\2023 SG AIP.xlsx"
df1 = pd.read_excel(file_path1, dtype={'Merchant No': str})
```

```python
df1
```

```python
df1.dtypes
```

```python
(df1['Payment Amount'] != df1['Total Amount']).sum()
```

#### Convert columns to proper data types

```python
df1['Transaction Date'] = pd.to_datetime(df1['Transaction Date'], format='%Y%m%d')
df1['Transaction Month'] = df1['Transaction Date'].dt.to_period('M')
df1['Merchant Name'] = df1['Merchant Name'].str.strip()
df1['Merchant No'] = df1['Merchant No'].str.strip()
df1['Card Organization'] = df1['Card Organization'].str.strip()
```

#### split columns into separate groups 

```python
df1 = df1.groupby(['Transaction Month', 'Merchant No', 'Merchant Name', 'Card Organization']).agg({'Payment Amount': 'sum', 'Settlement Amount': 'sum', 'Transaction Fee': 'sum', 'gst': 'sum', 'Transaction No': 'count'})
```

### SG 2024 data processing

```python
# Read the Excel file
file_path2 = r"C:\Users\Kitty Wong\OneDrive - BBMSL Limited\02.99 Intern\03 Interns\05 KW\Data\SG\2024 SG AIP.xlsx"
df2 = pd.read_excel(file_path2, dtype={'Merchant No': str})
```

```python
df2
```

#### remove rows

```python
df2 = df2[~df2['Transaction Type'].str.contains('Refund', na=False)]
```

```python
df2
```

#### Shift rows

```python
mask = df2['Card Organization'].astype(str).str.contains('000122|000096')
df2.loc[mask, 'Card Organization':] = df2.loc[mask, 'Card Organization':].shift(axis=1)
```

```python
df2.dtypes
```

```python
df2['Transaction Date'] = pd.to_datetime(df2['Transaction Date'], format='%Y%m%d')
df2['Transaction Month'] = df2['Transaction Date'].dt.to_period('M')
df2['Merchant Name'] = df2['Merchant Name'].str.strip()
df2['Merchant No'] = df2['Merchant No'].str.strip()
df2['Transaction Type'] = df2['Transaction Type'].str.strip()
df2['Card Organization'] = df2['Card Organization'].str.strip()
df2['Payment Amount'] = df2['Payment Amount'].astype(float)
df2['Settlement Amount'] = df2['Settlement Amount'].astype(float)
```

#### check & compare columns

```python
(df2['Payment Amount'] != df2['Total Amount']).sum()
```

#### fill in empty blanks in specific column

```python
df2.loc[(df2['Transaction Type'] == 'PAYNOW_SCANCODE') & (df2['Card Organization'] == ''), 'Card Organization'] = 'PAYNOW'
df2.loc[((df2['Transaction Type'] == 'Wechat customer persent QRcode Payment-Wechat C-QRC') | 
     (df2['Transaction Type'] == 'Wechat merchant present QRcode payment-Wechat M-QRC')) & (df2['Card Organization'] == ''), 'Card Organization'] = 'WECHATPAY'
```

```python
df2[['Transaction Type', 'Card Organization']]
```

```python
df2 = df2.groupby(['Transaction Month', 'Merchant No', 'Merchant Name', 'Card Organization']).agg({'Payment Amount': 'sum', 'Settlement Amount': 'sum', 'Transaction Fee': 'sum', 'gst': 'sum', 'Transaction No': 'count'})
```

```python
df2
```

### combination of 23 & 24 SG data file

```python
df = pd.concat([df1, df2])
```

```python
df.to_csv('SGcombination.csv')
```
